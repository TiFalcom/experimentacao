```python
import pandas as pd
import os
import sys
from pathlib import Path
from datetime import datetime

project_dir = Path('__main__').resolve().parents[1]
```

```python
df = pd.read_csv(os.path.join(project_dir, 'data', 'stress_hlp.txt'), delimiter=';', )
```

```python
for col in df.columns:
    df[col] = df[col].apply(lambda x: datetime.strptime(x.replace("'", ""), '%Y-%m-%d%H:%M:%S.%f'))
```

```python
col_anterior = None
for col in df.columns:
    if col_anterior is not None:
        df[f'diff_{col}'] = (df[col] - df[col_anterior]).dt.total_seconds()*100
    col_anterior = col
```

```python
codigo_fonte = '''    def transform(self, X : pd.DataFrame):\n        X = X[self.variaveis_utilizadas].reset_index(drop=True)\n        X_tmp = X.reset_index(drop=True)\n\n        X_tmp['valor_parcela'] = X_tmp['valor_transacao'] / X_tmp['quantidade_parcelas']\n\n        X_tmp['periodo_do_dia'] = X_tmp['data_e_hora'].str[0:2].map(self.map_horario)\n\n        colunas_criadas = [coluna for coluna in X_tmp.columns if coluna not in X.columns]\n        X_tmp = X.merge(X_tmp[colunas_criadas], right_index=True, left_index=True)\n        return X_tmp\n'''

linhas = codigo_fonte.split('\n')
novo_codigo_fonte = []

for linha in linhas:
    if linha.strip():
        if linha.strip().startswith(('return')):
            novo_codigo_fonte.append(linha.replace('\n', '').rstrip() + ', timestamps\n')
        else:
            novo_codigo_fonte.append(linha)
        if linha.strip().startswith(('def ')):
            novo_codigo_fonte.append('        timestamps = []')
        if not linha.strip().startswith(('class ', 'return', 'import', 'from')):
            novo_codigo_fonte.append('        ' + 'timestamps.append(datetime.now())\n')

novo_codigo_fonte = '\n'.join(novo_codigo_fonte)
```
