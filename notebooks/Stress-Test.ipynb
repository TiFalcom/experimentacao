```python
import pandas as pd
import os
import sys
from pathlib import Path
from datetime import datetime

project_dir = Path('__main__').resolve().parents[1]
```

```python
df = pd.read_csv(os.path.join(project_dir, 'data', 'stress', 'response.txt'), delimiter=';', )
```

```python
for col in df.columns[1:]:
    df[col] = df[col].apply(lambda x: datetime.strptime(x.replace("'", ""), '%Y-%m-%d%H:%M:%S.%f'))
```

```python
col_anterior = None
for col in df.columns[1:]:
    if col_anterior is not None:
        df[f'diff_{col}'] = (df[col] - df[col_anterior]).dt.total_seconds() * 100
    col_anterior = col
```
